```python
train1.isnull().sum()
```

```python
train1.drop(columns = 'Cabin', axis = 1, inplace = True)
```

```python
train1['Age'].fillna(train['Age'].mean(), inplace = True)
```

```python
train1.dropna(inplace = True)
```

```python
from numpy import int64
sex = pd.get_dummies(train1['Sex'], drop_first = True, dtype = int64)
```

```python
sex.head(3)
```

```python
embarked = pd.get_dummies(train1['Embarked'], drop_first = True, dtype = int64)
```

```python
pclass = pd.get_dummies(train1['Pclass'], drop_first = True, dtype = int64)
```

```python
train = 0
```

```python
train = pd.concat([train1, pclass, sex, embarked], axis = 1)
```

```python
train.head(3)
```

```python
train.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Ticket', 'Embarked'], axis = 1, inplace = True)
```

```python
train.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cols_to_scale = ['Age', 'Fare']
scaler.fit(train[cols_to_scale])

```

```python
train[cols_to_scale] = scaler.transform(train[cols_to_scale])

train.head(3)
```

```python
train.columns
```

```python
train.columns = train.columns.map(str)
```

```python
X_train = train.drop('Survived', axis = 1)
```

```python
Y_train = train['Survived']
```

```python
X_train.head(3)

#Y_train.head(3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
train.isnull().sum()
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred_train = model.predict(X_train)
```

```python
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(Y_train, y_pred_train)
```

```python
conf_matrix_df = pd.DataFrame(conf_matrix)
```

```python
conf_matrix_df
```

```python
conf_matrix_labels = pd.DataFrame(conf_matrix, columns = ['prognose dead', 'prognose alive'], index = ['fact dead', 'fact alive'])

```

```python
conf_matrix_labels
```
